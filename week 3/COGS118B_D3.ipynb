## K-Means

The K-means algorithm is elegantly simple yet remarkably powerful. At its core, the algorithm aims to partition a dataset into K distinct, non-overlapping subsets or clusters. It achieves this by minimizing the variance within each cluster, a method known as intra-cluster variance. The essence of K-means lies in identifying K centroids, one for each cluster, and assigning each data point to the nearest centroid, thereby creating K clusters based on proximity.
